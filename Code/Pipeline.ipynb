```python
from sklearn.neighbors import KNeighborsRegressor
import pandas as pd
import numpy as np

from main import import_dataset_from_file
```

```python
knns = []

xyxyxy = []
zzz = []

for i in range(5):
    print(f'Loading dataset #{i+1}')
    df = import_dataset_from_file(f'../Data/Map_{i+1}.txt')
    X, y = df[['x', 'y']].values, df['z'].values

    if i == 0:
        xyxyxy.append(X)
        zzz.append(y)

    print('kNN training')
    knn = KNeighborsRegressor(n_neighbors=3, weights='distance', metric='l2')
    knn.fit(X, y)
    knns.append(knn)
```

```python
points_df = import_dataset_from_file('../Data/Point_dataset.txt')
points_df
```

```python
cols = ['x', 'y', 'map1', 'map2', 'map3', 'map4', 'map5', 'F']
df = pd.DataFrame(columns=cols)
df[['x', 'y', 'F']] = points_df[['x', 'y', 'z']]
df
```

```python
for i, knn in enumerate(knns):
    xy = df[['x', 'y']].values.astype(np.float32)
    df[f'map{i+1}'] = knn.predict(xy)

df
```

```python
from sklearn.metrics import mean_squared_error


b = knns[4].predict(np.array(xyxyxy[0]))

mse = mean_squared_error(np.array(zzz[0]), b)
mse**0.5
```
